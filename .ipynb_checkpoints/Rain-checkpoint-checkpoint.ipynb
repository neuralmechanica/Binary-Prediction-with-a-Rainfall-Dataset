## Binning

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import KBinsDiscretizer
```

```python
columns_to_bin=['dewpoint', 'humidity', 'cloud', 'sunshine', 'DeltaTemp', 'wc', 'pressure',	'maxtemp',	'temparature',	'mintemp', 'winddirection',	'windspeed']
kbin = KBinsDiscretizer(n_bins=15, encode='ordinal', strategy='quantile')
trf = ColumnTransformer(
    transformers=[('bin', kbin, columns_to_bin)],
    remainder='passthrough',
    verbose_feature_names_out=False
)
```

```python

X_trf= trf.fit_transform(X)
test_trf=trf.transform(test)
```

```python
feature_names = trf.get_feature_names_out()

 #Construct DataFrames with accurate column names
X_transformed = pd.DataFrame(X_trf, columns=feature_names, index=X.index)
test_transformed = pd.DataFrame(test_trf, columns=feature_names, index=test.index)
```

```python
X_transformed.head()
```

## Feature Transforming

we do feature transform to make a feature normally distributed. Normally distributed dataset perform well in ML models

```python
import scipy.stats as stats
from sklearn.preprocessing import PowerTransformer
import statsmodels.api as sm
```

```python
trans = PowerTransformer()
X_scaled = trans.fit_transform(X_transformed)
test_scaled = trans.transform(test_transformed) ## when we do yeo-jhonson transformer , we don't need to do feature scaling. yeo-jhonson do feature scaling by default ##
```

```python
X_scaled = pd.DataFrame(X_scaled, columns = X_transformed.columns)
test_scaled = pd.DataFrame(test_scaled, columns = test_transformed.columns)
```

```python
X_scaled.head()
```

## Model Train and Evaluation

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
kfold = KFold(n_splits = 10, shuffle = True, random_state =0)
clf = DecisionTreeClassifier()
scoring = 'accuracy'
score = cross_val_score(clf,x_final, y_final, cv=kfold, n_jobs=1, scoring=scoring)
print(score)
round(np.mean(score)*100, 2)
```

```python
clf = RandomForestClassifier(n_estimators=100)
scoring = 'accuracy'
score = cross_val_score(clf, x_final, y_final, cv=kfold, n_jobs=1, scoring=scoring)
print(score)
round(np.mean(score)*100, 2)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
scoring = 'accuracy'
score = cross_val_score(clf,x_final, y_final, cv=kfold, n_jobs=1, scoring=scoring)
print(score)
round(np.mean(score)*100, 2)
```

```python
from xgboost import XGBClassifier
clf = XGBClassifier(use_label_encoder = False, eval_metrics = 'logloss')
scoring = 'accuracy'
score = cross_val_score(clf, x_final, y_final, cv=kfold, n_jobs=1, scoring=scoring)
print(score)
round(np.mean(score)*100, 2)
```

```python
from sklearn.svm import SVC
clf = SVC(kernel = 'rbf')
scoring = 'accuracy'
score = cross_val_score(clf, x_final, y_final, cv=kfold, n_jobs=1, scoring=scoring)
print(score)
round(np.mean(score)*100, 2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=10)
scoring = 'accuracy'
score = cross_val_score(clf, x_final, y_final, cv=kfold, n_jobs=1, scoring=scoring)
print(score)
round(np.mean(score)*100, 2)
```

# Ensemble Learning

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
mod1=DecisionTreeClassifier()
mod2=RandomForestClassifier(n_estimators=100)
mod3=LogisticRegression()
mod4=XGBClassifier()
mod5=SVC(probability=True)
mod6=KNeighborsClassifier()
```

```python
estimators = [('DT', mod1), ('RF', mod2), ('LR', mod3),('XGB', mod4), ('svc', mod5), ('knn', mod6)]
```

```python
# Testing all the models's accuracy individually
for estimator in estimators:
    x=cross_val_score(estimator[1], x_final, y_final, cv=10, scoring='accuracy')
    print(estimator[0], np.round(np.mean(x), 2))
```

### Voting Classifier

```python
from sklearn.ensemble import VotingClassifier
```

Hard Voting Classifier

```python
vc=VotingClassifier(estimators=estimators, voting='hard')
x=cross_val_score(vc, x_final, y_final, cv=10, scoring='accuracy')
print(np.round(np.mean(x), 2))
```

```python
vc.fit(x_final, y_final)
prediction = vc.predict(test_scaled)
submission = pd.DataFrame({'id':ID, 'rainfall':prediction})
submission.to_csv('HVC_full.csv', index=False)
```

soft Voting Classifier

```python
vc_soft=VotingClassifier(estimators=estimators, voting='soft')
x=cross_val_score(vc_soft, x_final, y_final, cv=10, scoring='accuracy')
print(np.round(np.mean(x), 2))
```

```python
vc_soft.fit(x_final, y_final)
prediction = vc_soft.predict(test_scaled)
submission = pd.DataFrame({'id':ID, 'rainfall':prediction})
submission.to_csv('SoftVC_full.csv', index=False)
```

weight Voting Classifier

```python
vc_w=VotingClassifier(estimators=estimators, voting='soft', weights=[3, 6, 1, 5, 2, 4])
x=cross_val_score(vc_w, x_final, y_final, cv=10, scoring='accuracy')
print(np.round(np.mean(x), 2))
```
